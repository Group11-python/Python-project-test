```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load all the csv files as data frames

df1 = pd.read_csv('Australia_Wine_Stats.csv')
df2 = pd.read_csv('Chile_Wine_Stats.csv')
df3 = pd.read_csv('Italy_Wine_Stats.csv')
df4 = pd.read_csv('France_Wine_Stats.csv')
df5 = pd.read_csv('New Zealand_Wine_Stats.csv')
df6 = pd.read_csv('Portugal_Wine_Stats.csv')
df7 = pd.read_csv('Spain_Wine_Stats.csv')
df8 = pd.read_csv('USA_Wine_Stats.csv')

# Combine all the data frames into a single data frame

wine_df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8], ignore_index=True)
```

```python
# First rows of the data frame
wine_df.head()
```

```python
# Information of the data frame
wine_df.info()
```

```python
# Remove duplicate records
wine_df.drop_duplicates()
```

```python
# Remove null records
wine_df.dropna()
```

```python
# Interquartile Range to remove outliers
def remove_outliers(wine_df, column):
    Q1 = wine_df[column].quantile(0.25)  # 25th percentile
    Q3 = wine_df[column].quantile(0.75)  # 75th percentile
    IQR = Q3 - Q1                   # Interquartile Range
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return wine_df[(wine_df[column] >= lower_bound) & (wine_df[column] <= upper_bound)]
```

```python
# Remove outliers
wine_df = remove_outliers(wine_df, 'Rating')
wine_df = remove_outliers(wine_df, 'Number of Ratings')
wine_df = remove_outliers(wine_df, 'Price')
wine_df = remove_outliers(wine_df, 'Alcohol content')
wine_df = remove_outliers(wine_df, 'Bold')
wine_df = remove_outliers(wine_df, 'Tannin')
wine_df = remove_outliers(wine_df, 'Sweet')
wine_df = remove_outliers(wine_df, 'Acidic')
```

```python
# First rows of the data set after removing outliers
wine_df.head()
```

```python
# Data points that lie far from the bulk of the data (Deviated) are the outliers. These are not errors, 
# but rather rare occurrences that are far from the central tendency.
```

```python
# Extract the country from the 'Region' column
wine_df["Country"] = wine_df["Region"].apply(lambda x: x.split(" / ")[0])
```

```python
# Function to extract country region
def extract_country_region(region):
    parts = region.split('/')
    if len(parts) > 1:
        return parts[1].strip()  # Return the second part as country region
    return None  # Handle cases where region data might be malformed

# Add the Country_region column
wine_df['Country_region'] = wine_df['Region'].apply(extract_country_region)
```

```python
# Details of the data set after extracting from the region column
wine_df.info()
```

```python
# Define the unique food items (you can extend this list as needed)
food_items = ['Beef','Lamb', 'Poultry', 'Pasta']

# Create new columns for each food item
for food in food_items:
    wine_df[food] = wine_df['Food pairings'].apply(lambda x: True if food in x else False)

# Display the updated DataFrame
print(wine_df)
```

```python
# Remove irrelevant columns
wine_df = wine_df.drop(columns=['Region', 'Grapes'])
```
